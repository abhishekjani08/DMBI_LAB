```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv('House_Rent.csv')
```

```python
backup_df = df
```

```python
records = []
for i in range (0,2000):
 records.append([str(df.values[i,j]) for j in range(0,7)])
```

```python
# Build Apriori model
association_rules = apriori(records, min_support = 0.003, min_confidence = 1)
association_results = list(association_rules)
```
