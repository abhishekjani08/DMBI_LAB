```python
#Part 5: How to create plots (Histogram, Scatter, Box Plot)?
```

```python
#Plot Histogram
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("House_Rent.csv")
#Plots in matplotlib reside within a figure object, use plt.figure to create new figure
fig=plt.figure()
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.hist(df['City'],bins = 5)
#Labels and Tit
plt.title('HOUSE RENT')
plt.xlabel('Cities')
plt.ylabel('Rent')
plt.show()
```

```python
#Scatter plot
```

```python
#Plots in matplotlib reside within a figure object, use plt.figure to create new figure
fig=plt.figure()
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.scatter(df['City'],df['Rent'])
#Labels and Tit
plt.title('HOUSE RENT')
plt.xlabel('Cities')
plt.ylabel('Rent')
plt.show()
```
